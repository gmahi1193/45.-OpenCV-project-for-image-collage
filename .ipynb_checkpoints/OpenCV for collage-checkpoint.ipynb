### Import libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

### Check for current working directory

```python
os.getcwd()
```

### Loading images using OpenCV

```python
image_1 = cv2.imread('top_left.jpg')
image_2 = cv2.imread('top_right.jpg')
image_3 = cv2.imread('center.jpeg')
image_4 = cv2.imread('bottom_left.jpg')
image_5 = cv2.imread('bottom_right.jpg')
```

### Converting image color

```python
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
image_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB)
image_3 = cv2.cvtColor(image_3, cv2.COLOR_BGR2RGB)
image_4 = cv2.cvtColor(image_4, cv2.COLOR_BGR2RGB)
image_5 = cv2.cvtColor(image_5, cv2.COLOR_BGR2RGB)
```

### Resizing images for collage

```python
width, height = 300,300
image_1 = cv2.resize(image_1,(width,height))
image_2 = cv2.resize(image_2,(width,height))
image_3 = cv2.resize(image_3,(200,200))
image_4 = cv2.resize(image_4,(width,height))
image_5 = cv2.resize(image_5,(width,height))
```

### Stacking of images

```python
photo1 = np.hstack([image_1,image_2])
photo2 = np.hstack([image_4,image_5])
photo3 = np.vstack([photo1,photo2])

# cv2.waitKey()
# cv2.destroyAllWindows()
```

```python
photo3[200:400,200:400,:] = image_3
```

### Visualizing image

```python
plt.figure(figsize = [10,10])
plt.title('Pokemon Collage')
plt.imshow(photo3)
plt.axis('off')
plt.show()
```
